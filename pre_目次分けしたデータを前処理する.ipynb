```python
import subprocess
import zenhan
import os
import sys
import re
import unicodedata
import argparse
import pandas as pd
import pickle
import json
import tqdm
```

```python
#企業リストを取得
company_list = pd.read_csv("../data/new2/df_achieve/company_list.csv",index_col=0)
company_list=company_list.drop(14350)
company_list=company_list.drop(52080)
company_list=company_list.drop(95080)
```

```python
def normalize(text):
    text = text.replace('．','')
    text = text.replace('.','')
    text = re.sub(r'＜[^＜|^＞]*＞', '', text)
    text = re.sub(r'【[^【|^】]*】', '', text)
    text = re.sub(r'［[^［|^］]*］', '', text)
    
    text = text.replace('『','「')
    text = text.replace('』','」')
#     text = text.replace('【','「')
#     text = text.replace('】','」')
    text = text.replace('△','')
    text = re.sub(r'\d+', '0', text)
    text = re.sub(r'（', '(', text)
    text = re.sub(r'）', ')', text)
    text = re.sub(r'，', ',', text)
    
    return(text)
```

```python
for company in tqdm.tqdm(company_list.index):
    try:
        df_path = '../data/new2/df_achieve/this/{}.csv'.format(company)
        #dfを呼び出し
        df = pd.read_csv(df_path,index_col=0)

        dic = '../data/text_data_ver2/text_data_sep_pre/'+str(company)
        #フォルダを作成する
        try:
            os.makedirs(dic)
        except FileExistsError:
            pass

        #企業の決算短信を一つ一つ見る
        for code in df['pdf_code']:
            json_path = '../data/text_data_ver2/text_data_sep/'+str(company)+'/'+str(code)+'.json'
            out_path = '../data/text_data_ver2/text_data_sep_pre/'+str(company)+'/'+str(code)+'.json'
            json_open = open(json_path, 'r')
            json_load = json.load(json_open)
            json_open.close()

            #決算短信の段落ごとにテキストの処理をする
            new_dic={'text':[]}
            for i,paragraph in enumerate(json_load["text"]):
                kari_dic=paragraph
                #段落ごとに文章を前処理して，改行コードで文章を分ける
                text = normalize(kari_dic["text"])
                text_list=text.split('\n')

                #文章として成立していない文章は削除
                sentence_list=[]
                for text in text_list:
                    if len(text) <3:
                        pass
                    else:
                        sentence_list.append(text)

                kari_dic["text"]=sentence_list
                new_dic["text"].append(kari_dic)


            #保存
            json_file = open(out_path, 'w')
            json.dump(new_dic, json_file, ensure_ascii=False)
            json_file.close()
    except:
        print(company)
```

```python
# テキストの取得
#企業リストを取得
company_list = pd.read_csv("../data/new2/df_achieve/company_list.csv",index_col=0)
company_list=company_list.drop(14350)
company_list=company_list.drop(52080)
company_list=company_list.drop(95080)
```
