```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import math
import csv

#選択行削除
import re
from xml.sax.saxutils import unescape
```

```python
def sentence(sentence):
    if sentence is None:
        return ''

    # 特殊文字デコード
    sentence = unescape(sentence)

    # URL削除
    sentence = re.sub(r'(https?)(:\/\/[-_.!~*\'()a-zA-Z0-9;\/?:\@&=+\$,%#]+)', "", sentence)

    # 「！」はすべて半角単！に
    sentence = re.sub(r'！+', "!", sentence)
    sentence = re.sub(r'!+', "!", sentence)

    # 「？」はすべて半角単？に
    sentence = re.sub(r'？+', "?", sentence)
    sentence = re.sub(r'\?+', "?", sentence)
    
    #．はすべて半角に
    sentence = re.sub(r'．', ".", sentence)
    sentence = re.sub(r'，', ",", sentence)

    # 「（」「）」はすべて半角に
    sentence = sentence.replace("（", "(")
    sentence = sentence.replace("）", ")")

    # 「！」「？」を削除
    sentence = sentence.replace("!", "")
    sentence = sentence.replace("?", "")

    # 全角空白削除
    sentence = sentence.replace("ㅤ", "")
    # スペースすべて削除
    sentence = sentence.replace(" ", "")

    # 余白削除
    sentence = sentence.strip()
    
    #数字はすべて0にする
    sentence = re.sub(r'\d+', '0',sentence)
    sentence = re.sub(r'0,0', '0',sentence)
    sentence = re.sub(r'0.0', '0',sentence)
    
    #()で囲まれている文字を削除する
    #sentence = sentence.replace(r"(.*)", "")
    sentence = re.sub(r'\([^)]*\)', '', sentence)
    
    #削除する単語　追加
    sentence = re.sub(r'【',"",sentence)
    sentence = re.sub(r'】',"",sentence)
    #sentence = re.sub(',','',sentence)
    #sentence = re.sub('．','',sentence)
    #sentence = re.sub('，','',sentence)
    sentence = re.sub('「','',sentence)
    sentence = re.sub('」','',sentence)
    sentence = re.sub('・','',sentence)
    sentence = re.sub('@','',sentence)
    sentence = re.sub('/','',sentence)
    sentence = re.sub('…','',sentence)
    sentence = re.sub('：','',sentence)
    sentence = re.sub(':','',sentence)
    sentence = re.sub('●','',sentence)
    sentence = re.sub('◆','',sentence)
    sentence = re.sub('★','',sentence)
    sentence = re.sub('■','',sentence)
    sentence = sentence.replace('(','')
    sentence = sentence.replace(')','')
    sentence = re.sub('△','',sentence)
    sentence = re.sub('『','',sentence)
    sentence = re.sub('』','',sentence)
    sentence = re.sub('※','',sentence)
    
    return sentence
```

```python
for i in range(0,1):
    for j in range(1):
        out='file/out/company{}/{}.txt'.format(i+1,j+1)
```

```python
f = open(out, 'r', encoding='UTF-8')
data = f.read()
data = sentence(data)
with open('test.txt','w') as out:
    out.write(data)
f.close()
```
